```python
original_db['DAY'] = pd.to_datetime(original_db['CRASH DATE']).dt.day
original_db['MONTH'] = pd.to_datetime(original_db['CRASH DATE']).dt.month
original_db['YEAR'] = pd.to_datetime(original_db['CRASH DATE']).dt.year

original_db['HOUR'] = pd.to_datetime(original_db['CRASH TIME']).dt.hour
original_db['MINUTE'] = pd.to_datetime(original_db['CRASH TIME']).dt.minute



added_db['DAY'] = pd.to_datetime(added_db['CRASH DATE']).dt.day
added_db['MONTH'] = pd.to_datetime(added_db['CRASH DATE']).dt.month
added_db['YEAR'] = pd.to_datetime(added_db['CRASH DATE']).dt.year

added_db['HOUR'] = pd.to_datetime(added_db['CRASH TIME']).dt.hour
added_db['MINUTE'] = pd.to_datetime(added_db['CRASH TIME']).dt.minute
```

```python
road_db["LATITUDE"] = "";
road_db["LONGITUDE"] = "";

for i in road_db.index:
    txt = road_db.at[i, 'the_geom']

    lat_long = txt.split("((")[1].split(",")[0].split(" ")
    long = lat_long[0]
    lat = lat_long[1]
    
    road_db.at[i, 'LATITUDE'] = lat
    road_db.at[i, 'LONGITUDE'] = lat
```

```python
added_db.drop(added_db.columns.difference(["LATITUDE","LONGITUDE","BOROUGH","ZIP CODE","ON STREET NAME","STREETWIDT","RW_TYPE","temp","dwpt","rhum","prcp","wdir","wspd","pres","DAY","MONTH","YEAR","HOUR","MINUTE"]), 1, inplace=True)
original_db.drop(original_db.columns.difference(["LATITUDE","LONGITUDE","BOROUGH","ZIP CODE","ON STREET NAME","STREETWIDT","RW_TYPE","temp","dwpt","rhum","prcp","wdir","wspd","pres","DAY","MONTH","YEAR","HOUR","MINUTE"]), 1, inplace=True)

road_db.drop(road_db.columns.difference(["LATITUDE","LONGITUDE","L_ZIP","ST_LABEL","BOROCODE", "ST_WIDTH", "RW_TYPE"]), 1, inplace=True)
```

```python
new_negatives = added_db.copy()
```

```python
import numpy as np
```

```python
replace_rows = np.random.choice(added_db.index, int(added_db.shape[0]/2), replace=False)

for count, added_row in enumerate(replace_rows):
    road_row = count % road_db.shape[0]
    
    new_negatives.at[added_row, 'ZIP CODE']= road_db.at[road_row, 'L_ZIP']
    new_negatives.at[added_row, 'ON STREET NAME']= road_db.at[road_row, 'ST_LABEL']
    new_negatives.at[added_row, 'BOROUGH']= road_db.at[road_row, 'BOROCODE']
    new_negatives.at[added_row, 'STREETWIDT']= road_db.at[road_row, 'ST_WIDTH']
    new_negatives.at[added_row, 'RW_TYPE']= road_db.at[road_row, 'RW_TYPE']
    new_negatives.at[added_row, 'LATITUDE']= road_db.at[road_row, 'LATITUDE']
    new_negatives.at[added_row, 'LONGITUDE']= road_db.at[road_row, 'LONGITUDE']
```

```python
new_head = new_negatives.head()
new_head
```

```python
new_negatives['CRASH'] = 0
added_db['CRASH'] = 0
original_db['CRASH'] = 1
```

```python
new_head = new_negatives.head()
new_head
```

```python
combined_df = pd.concat([original_db, new_negatives], ignore_index=True, sort=False)
```

```python
milestone_combined_df = pd.concat([original_db, added_db], ignore_index=True, sort=False)
```
